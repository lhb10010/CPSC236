````python
### CODE HERE ###
import csv


def readTemplate():

    template = ""
    
    try:
        with open('email_template.txt', 'r') as file:
            template = file.read()
    except:
        print("Failed to open template\n")
        exit(1)
    return template


def readEmails():

    rows = []

    try:
    
        with open('email.csv', 'r') as file:
            csv_reader = csv.reader(file)
            for row in csv_reader:
                rows.append(row)

    except:
        print("Failed to open emails\n")
        exit(1)

    return rows
    

def printFromattedEmails(template, emailRows):

    for row in emailRows:
        templateCopy = template
        formattedEmail = templateCopy.replace("{email}", row[2].lower()).replace("{first_name}", row[0].title())
        print(formattedEmail)



if __name__ == "__main__":
    print("Email Creator\n")
    printFromattedEmails(readTemplate(), readEmails())

````

## 10.2 Pig Latin Translator

Create a program that translates English to Pig Latin.

### Console:
```powershell
Pig Latin Translator

Enter text: Tis but a scratch. 
English:	tis but a scratch
Pig Latin:	istay utbay away atchscray 

Continue? (y/n): y

Enter text: We are the knights who say nee!
English:	we are the knights who say nee
Pig Latin:	eway areway ethay ightsknay owhay aysay eenay 

Continue? (y/n): n

Bye!
```

### Specifications:
- Convert the English to lowercase before translating.
- Remove the following punctuation characters before translating `! , ; : - . ?`
- Assume that words are separated from each other by a single space.
- If the word starts with a vowel, just add *way* to the end of the word.
- If the word starts with a consonant, move all of the consonants that appear before the first vowel to the end of the word, then add *ay* to the end of the word.

### Note:
- There are no official rules for Pig Latin. Most people agree on how words that begin with consonants are translated, but there are many different ways to handle words that begin with vowels.


````python
### CODE HERE ###
def printEnglish(words):
    
    print("English: ", end="")
    for word in words:
        print(word, end=" ")
    print("")
        
        
def printPigLatin(words):

    print("Pig Latin: ", end="")
    for word in words:
        if word.startswith("a") or word.startswith("e") or word.startswith("i") or word.startswith("o") or word.startswith("u"):
            print(word + "way", end=" ")
        else:
            removeFromBeggining = 0
            lettersToMove = ""
            for char in word:
                
                if not char in ["a", "e", "i", "o", "u"]:
                    lettersToMove = lettersToMove + char
                    removeFromBeggining = removeFromBeggining + 1

                else:
                    break
                    
            word = word[removeFromBeggining:]
            word = word + lettersToMove
            word = word + "ay"
            print(word, end=" ")
            
    print("")
    

if __name__ == "__main__":
    
    print("Pig Latin Translator")
    
    while True:
    
        inp = input("\nEnter text: ")

        invalidChars = "!,;:-.?"
        
        for char in invalidChars:
            inp = inp.replace(char, "")

        words = inp.split()
    
        for i in range(0, len(words)):
            words[i] = words[i].lower()
            
        printEnglish(words)
        printPigLatin(words)
        
        inp = input("\nContinue? (y/n): ")
        if not inp.lower()[0] == "y":
            break
        
            
````
