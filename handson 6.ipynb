## 12.1 Champion Counter
Create a program that reads a text file that contains a list of FIFA World Cup champions and determines the country that has won the most championships.

### Console:
```powershell
Country	        Wins	Years	
=======	        ====	=====	
Argentina       2       1978, 1986
Brazil	        5       1958, 1962, 1970, 1994, 2002
England	        1       1966		
France	        1       1998		
Germany	        4       1954, 1974, 1990, 2014
Italy	        4       1934, 1938, 1982, 2006
Spain	        1       2010
Uruguay	        2       1930, 1950
```

### Specifications:
- Use the attached text file named `world_cup_champions.txt` that contains data like this: 
```powershell
Year,Country,Coach,Captain 
1930,Uruguay,Alberto Suppici,Jos√© Nasazzi 
1934,Italy,Vittorio Pozzo,Gianpiero Combi 
1938,Italy,Vittorio Pozzo,Giuseppe Meazza
...
```
- When the program starts, it should read the text file and use a dictionary to store the required data using the name of each country that has won the World Cup as the key.
- The program should compile the data shown above and display the countries alphabetically.



````python
### CODE HERE ###

def readWorldCupDataFromFile(filename):
    
    file = open(filename, 'r')
    
    rows = file.readlines()
    rows.pop(0)
    dataDict = {}
    
    for row in rows:
        rowList = row.split(",")
        if rowList[1] in dataDict.keys():
            countryList = dataDict[rowList[1]]
            countryList[1] = countryList[1] + 1
            countryList[2].append(int(rowList[0]))
            
        else:
            yearList = []
            yearList.append(rowList[0])
            countryList = []
            countryList.append(rowList[1])
            countryList.append(1)
            countryList.append(yearList)
            dataDict[countryList[0]] = countryList
            
    file.close()
    
    return dataDict


def displayDataDict(dataDict):
    
    print("Country\t\tWins\tYears\n=======\t\t====\t=====")
     
    keyStringList = []
    for key in dataDict.keys():
        keyStringList.append(str(key))
    
    keyStringList.sort()
    
    for key in keyStringList:
        
        entry = dataDict[key]
        
        print(entry[0] + "   \t" + str(entry[1]) + "\t", end = "")
        first = True
        for year in entry[2]:
            if not first:
                print(", ", end = "")
            else:
                first = False
            print(str(year), end = "")
    
        print("")
        
        
if __name__ == "__main__":
    
    displayDataDict(readWorldCupDataFromFile("world_cup_champions.txt"))
````

## 12.2 Monthly Sales
Create a program that allows you to view and edit the sales amounts for each month of the current year.

### Console:
```powershell

COMMAND MENU
view	- View sales for specified month 
edit	- Edit sales for specified month 
totals - View sales summary for year 
exit	- Exit program

Command: view
Three-letter Month: jan
Sales amount for Jan is 14,317.00.

Command: edit
Three-letter Month: jan 
Sales Amount: 15293
Sales amount for Jan is 15,293.00.

Command: totals
Yearly total: 67,855.00
Monthly average: 5,654.58

Command: view
Three-letter Month: july 
Invalid three-letter month.

Command: exit Bye!
```

### Specifications:
- Use the attached text file named `monthly_sales.txt` that consists of rows that contain three-letter abbreviations for the month and the monthly sales.
- The program should read the file and store the sales data for each month in a dictionary with the month abbreviation as the key for each item.
- Whenever the sales data is edited, the program should write the changed data to the text file.




````python
### CODE HERE ###

def readMonthlySalesFile(filename):
    
    file = open(filename, 'r')
    
    rows = file.readlines()
    
    dataDict = {}
    
    for row in rows:
        
        if not row == "":
            lineData = row.split("\t")
            dataDict[lineData[0].lower()] = float(lineData[1])
        
    file.close()
    
    return dataDict


def saveDataChanges(dataDict, filename):
    
    with open(filename, 'w') as file:
    
        for key in dataDict.keys():
            
            file.write(key + "\t" + str(dataDict[key]) + "\n")


def displayMenu():
    
    print("COMMAND MENU\nview    - View sales for specified month \nedit    - Edit sales for specified month \ntotals - View sales summary for year \nexit    - Exit program")


def edit(dataDict):
        
    inp = input("Three-letter Month: ")
    
    if not inp.lower() in dataDict.keys():
        print("Not a valid month\n")
        return
        
    amount = float(input("Sales Amount: "))
    dataDict[inp] = amount
    saveDataChanges(dataDict, "monthly_sales.txt")
    print("Sales amount for " + inp + " is " + str(dataDict[inp]) + "\n")
        
        
def view(dataDict):
    
    inp = input("Three-letter Month: ")
    
    if not inp.lower() in dataDict.keys():
        print("Not a valid month\n")
        return
    
    print("Sales amount for " + inp + " is " + str(dataDict[inp]) + "\n")
    
    
def totals(dataDict):
    
    total = 0
    
    for key in dataDict.keys():
        
        total = dataDict[key] + total
        
    average = total / 12
    print("Yearly total: " + str(total))
    print("Monthly average: " + str(average) + "\n")
    
    
if __name__ == "__main__":
    
    dataDict = readMonthlySalesFile("monthly_sales.txt")
    
    while True:
        
        displayMenu()
        inp = input("Command: ")
        
        if inp.lower() == "view":
            view(dataDict)
        elif inp.lower() == "edit":
            edit(dataDict)
        elif inp.lower() == "totals":
            totals(dataDict)
        elif inp.lower() == "exit":
            break
        else:
            print("Invalid input, try again\n")
        
````
